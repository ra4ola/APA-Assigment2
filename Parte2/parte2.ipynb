```python
#Preparação do dataset
#só quero imagens com carros

######base só faz numa imagem
def imagens_filtradas(path):


#abrir ficheiro i
  f=open(path, 'r')

  #ler linhas
  count = 0

  #ciclo for
  for linha in f:
      #print("Line{}: {}".format(count, linha.strip()))

      obs=float(linha[0:3])

      #ver que tem carros
      if(obs==0.0):
        #se sim guardar i na tabela valid_l
        #print("Classe carro")
        count += 1
      else:
        count=count
        #print("Outra classe qq")

  if(count==0):
    #print("Esta imagem nao tem carros")

    f.close()

    return False
  else:
    #print("A imagem tem {} carros".format(count))
      #fechar ficheiro i
    f.close()

    return True

```

```python
def read_lab(loc):
  #labels com carros
  valid_l=[]

  #dataset tem 500 imagens
  for i in range(500):
    imagem=str(i).zfill(6)
    imagem=loc+imagem+".txt"
    #print(imagem)


    if(imagens_filtradas(imagem)):
      valid_l.append(i)

  #print(valid_l)
  return valid_l
```

```python
def del_im_l(valid_l,dim):
  n=len(valid_l)

  import os

  loc_t='/content/gdrive/MyDrive/YOLO/Dataset/train/'
  loc_v='/content/gdrive/MyDrive/YOLO/Dataset/valid/'

  count=0
  for i in range(500):
    if(i not in valid_l):
      #delete file from train file
      print("Train delete ",i)
      l_in=loc_t+"labels/"+str(i).zfill(6)+".txt"
      i_in=loc_t+"Imgs/"+str(i).zfill(6)+".png"

      os.remove(l_in)
      os.remove(i_in)

      #delete file from valid file
      print("Valid delete ",i)
      l_in=loc_v+"labels/"+str(i).zfill(6)+".txt"
      i_in=loc_v+"Imgs/"+str(i).zfill(6)+".png"

      os.remove(l_in)
      os.remove(i_in)
    else:
      count=count+1;
      if(count>=dim):
        #delete file from train file
        print("-Train delete ",i)
        l_in=loc_t+"labels/"+str(i).zfill(6)+".txt"
        i_in=loc_t+"Imgs/"+str(i).zfill(6)+".png"

        os.remove(l_in)
        os.remove(i_in)

      else:
        #delete file from valid file
        print("-Valid delete ",i)
        l_in=loc_v+"labels/"+str(i).zfill(6)+".txt"
        i_in=loc_v+"Imgs/"+str(i).zfill(6)+".png"

        os.remove(l_in)
        os.remove(i_in)

```
